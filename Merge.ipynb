```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("C:/NLP_Project_1/messages_4.csv")
df.head()
```

```python
print(df['0'][4].split())
```

```python
def merge_columns(row):
    merged_values = []
    for col in row:
        if pd.notnull(col) and col != '[]' and not re.search(r'Jumpers J5562 and J5563' , col) :
            merged_values.append(col)
    return ' '.join(merged_values)
```

```python
df['Merged'] = df.iloc[:, 0:12].apply(merge_columns, axis=1)
df
```

```python
df.to_csv("C:/NLP_Project_1/cleaned.csv")
```
